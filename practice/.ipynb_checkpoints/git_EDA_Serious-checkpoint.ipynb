```python
import pandas as pd
%cd '~/git/nogitfile'
!ls
ori_df = pd.read_csv('train.csv')

### 看有幾個column
print(f'-共有{ori_df.shape[0]}筆資料')
#print('-共有{0}筆資料'.format(df.shape[0]))
print('-分布於%d欄'%ori_df.shape[1])
import numpy as np
def timeSinCosTrans(df):

    Time = df.loc[:,'loctm'].values
    # // 是整除的意思
    s, m, h = [(Time//100**(i)%100) for i in range(3)]
    Time = h*3600+m*60+s

    dfTime = pd.DataFrame()
    dfTime.loc[:,'TIME_SIN'] = np.sin(2*np.pi*Time/(24*60*60))
    dfTime.loc[:,'TIME_COS'] = np.cos(2*np.pi*Time/(24*60*60))
    dfTime.loc[:,'time'] = Time

    #df.drop(columns = 'loctm', inplace =True)
    df = pd.concat([df, dfTime], axis = 1)

    return df
```

### 相關性分析參考 
http://www.r-web.com.tw/guider/1/section_B_example_Q1.php

在探討兩變數間的相關性時，依據變數型態的不同，分析的方法也會不同。通常變數區分為連續型變數與類別型變數，變數的型態主要是由資料的 **測量尺度(scale of measurement)** 來決定，統計學中概分為 __四種測量尺度__ ，名目尺度(nominal scale)、順序尺度(ordinal scale)、區間尺度(interval scale)與比例尺度(ratio scale)。

- 名目尺度：數字只是一個代號，並不具測量的意義，如以0代表女性，1代表男性，0與1之間無法運算。
- 順序尺度：可依相對大小來排序，並給予對應的數字，此數字僅代表測量值的差異，但無法表示出差異的強度；如顧客滿意度區分為1至4，1為最差4為最佳。
- 區間尺度：測量值的數字可表示出差異的距離，但此數字無法計算比值，如時間的差異，9點與10點有1個小時的差距，但無法說出倍數的關係。
- 比例尺度：此種尺度的數字可以做任何的計算，0是一個絕對的原點，各種體積與面積的計算皆屬於此種尺度。

上述四種尺度中:
- 名目與順序尺度的資料型態為類別型變數
- 而區間與比例尺度的資料型態則為連續型變數
- 當兩變數皆為連續型變數時，適合以皮爾森(Pearson)相關分析來計算其相關係數之高低
- 當兩變數皆為類別型變數時，資料為順序關係可以斯皮爾曼等級相關(Spearman rank correlation)來了解變數間的相關性
- 當資料為順序尺度時則適合:
    1. 肯德爾等級相關性(Kendall tau)檢定
- 當資料為名目尺度時則適合:
    1. 費雪(R. A. Fisher)的exact檢定分析
    2. 卡方的獨立性(independence)與齊一性(homogeneity)檢定

## feature 'stocn'的工程
目標是兩種假設:
1. 在該國的交易總數與詐騙總數有關係
2. 在訓練資料中被詐騙比率較高的國家，在測試時也會較高

##### 做完以上假設後，valid f1提高 0.66 -> 0.68

#### Note: 不要輕易刪除資料
即使已經將資料轉換成更有用的資料，也不一定要把原來的資料消除，原因是資料也許另外有超越我們假設的特徵或關聯。

```python
lis_des = []
for i in lis_stocn:
    a = new_df[new_df['stocn'] == i]['fraud_ind'].values.sum()
    b = a/len(new_df[new_df['stocn'] == i])
    lis_des.append([i, a, b])

    
df = pd.DataFrame(lis_des)
df = df.loc[:][2]
#觀察機率分布
print(df.describe())

#根據機率Rank
# df.sort_values(ascending=False).index
count = 1
for i in df.sort_values(ascending=False).index:
    #print(lis_des[i])
    lis_des[i][2] = count
    #print(count)
    count += 1

#根據交易次數Rank
df = pd.DataFrame(lis_des)
df = df.loc[:][1]

count = 1
for i in df.sort_values(ascending=False).index:
    #print(lis_des[i])
    lis_des[i][1] = count
    #print(count)
    #print(lis_des[i])
    count += 1
```

```python
df = new_df['stocn'].copy()

# print(len(set(df)))
lis_value_index = []
for i in lis_des:
    lis_value_index.append([i,df[df == i[0]].index])
#lis_value_index[0][0][2]

for i in lis_value_index:
    df[i[1]] = i[0][2]
    
print(len(set(df)))
probRank_stocn = pd.DataFrame(df)
probRank_stocn = probRank_stocn.rename(columns={"stocn": "probRank_stocn"})

df = new_df['stocn'].copy()

for i in lis_value_index:
    df[i[1]] = i[0][1]
print(len(set(df)))

amountRank_stocn = pd.DataFrame(df)
amountRank_stocn = amountRank_stocn.rename(columns={"stocn": "amountRank_stocn"})
```

## feature 'csmcu'的工程 消費地幣別
目標假設包括:
1. 消費幣別與詐騙是有相關的。(原來的資料集)
2. 在訓練資料中被詐騙比率較高的幣別，在測試時也會較高
3. 在訓練資料中，被詐騙次數較高的幣別，在測試時也會較高

加入上述資料後發現預測率有明顯下降(valid f1 0.7 > 0.68)。判斷可能包括:

    1. 資料過多
    2. 模型複雜度不夠

>基於第一點，在加入'csmcu'特徵工程的前提下

    1. 移除'stocn'的特徵工程。重新測試。valid f1 0.71076747 0.71014493 0.70390754
    2. 移除'scity'的特徵工程。重新測試。valid f1 0.6975945 0.67496382 0.65219941
    3. 不加入'csmcu'特徵工程的情況中測試三次valid f1分別為 0.71405049 0.6893617 0.71047459

看起來加入'csmcu'特徵工程，在當前模型複雜度下，似乎會使正確率下降。暫時不考慮加入該特徵工程。
>基於第二點，提高模型複雜度，樹的 max-depth修改 (4->8)

    1. 保留所有資料與特徵工程進行3次測試，valid f1 {0.76615553, 0.75006683, 0.75862069}
    2. 移除 'stocn'特徵工程進行3次測試，valid f1 {0.75335841, 0.75995694, 0.76038856}
    3. 移除 'scity'特徵工程進行3次測試，valid f1 {0.73084371, 0.73925339, 0.75116343}
    4. 移除 'coman'特徵工程進行3次測試, valid f1{0.76088789, 0.76918991, 0.76393001}

依據上述測試判斷，coman在上述的特徵工程表現不佳，會導致預測效果降低。

- 新增特徵工程，one-hot編碼 
    - valid f1 {0.75430445, 0.74079916, 0.7482537 } 效果不彰

#### Notes: 資料的大小排列會影響決策樹的分類
由於我們使用的模型是xgboost，也就是一種決策樹模型，他會依據變數的數值來進行分類和學習，基本上是基於 ">" or "<" 的關係來表示。所以，盡可能的讓相似資料聚集在相似的數字附近會更有利分類，或者，使用one-hot編碼來進行資料處理。

```python
import xgboost as xgb
# from sklearn.model_selection import GridSearchCV
# parameters = { 'max_depth':[12, 16]}
model = xgb.XGBClassifier(max_depth= '8', learning_rate=.05, n_estimators = 1000\
                          ,tree_method = 'gpu_hist', predictor = 'gpu_predictor')
# clf = GridSearchCV(model, parameters, cv=5, scoring ='f1').fit\
#     (X=X_train,y=y_train, eval_set=[(X_train, y_train), (X_valid, y_valid)])

model.fit(X=X_train,y=y_train, eval_set=[(X_train, y_train), (X_valid, y_valid)], verbose = True)
```

```python
from sklearn.metrics import f1_score
# 如果沒有做one hot 全資料

y_true = y_valid.values
y_pred = model.predict(X_valid)

f1_score(y_true, y_pred, average=None)

```
