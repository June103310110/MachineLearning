## 把亂碼欄位換成Label Encoding

```python
strange_lis = ['source', 'browser', 'country', 'user_id', 'device_id']
data2 = df.copy()

from sklearn import preprocessing

def label_enc(data):
    le = preprocessing.LabelEncoder()
    le.fit(data)
    arr = le.transform(data)
    data = pd.DataFrame(arr)
    return data

for i in strange_lis:
    data2[i] = label_enc(data2[i])
```

```python
data2.head(5)
```

## 調整一下index的值 方便等一下作時間轉換

```python
data2.reset_index(drop=True, inplace =True)
```

```python
col = ['signup_time', 'purchase_time']
def new_col_of_time(data2, col):
    lis_hour = []
    lis_month =[]
    lis_day =[]
    for i in range(len(data2)):
        hour = int(data2[col][i][-8:-6])
        month = int(data2[col][i][5:7])
        day = int(data2[col][i][8:10])
        lis_hour.append(hour)
        lis_month.append(month)
        lis_day.append(day)

    b = col+'_hour'
    data2[b] = pd.DataFrame(lis_hour)
    
    b = col+'_month'
    data2[b] = pd.DataFrame(lis_month)
    
    b = col+'_day'
    data2[b] = pd.DataFrame(lis_day) 

new_col_of_time(data2, col[0]) 
new_col_of_time(data2, col[1]) 

data2.pop(col[0])
data2.pop(col[1])

```

## 確認沒有不合XGB規範的資料格式

```python
X_train.select_dtypes(include='object').any()
```

```python
from sklearn.model_selection import train_test_split
from sklearn import metrics
from xgboost.sklearn import XGBClassifier
X_train, X_valid, y_train, y_valid = train_test_split(
        X_train, y_train, test_size=0.1, random_state=42, stratify=y_train)
```

```python
from sklearn import metrics
from xgboost.sklearn import XGBClassifier

model = XGBClassifier(max_depth= 8, eta = 0.3, n_estimators = 1000, learning_rate = 0.005,
                          tree_method = 'gpu_hist', predictor = 'gpu_predictor',
                          random_state =12, scale_pos_weight = 9.738255033557047,
                          reg_lambda = 1, reg_alpha = 0
                         )

# 模型 訓練

model.fit(X=X_train,y=y_train, eval_set=[(X_train, y_train), (X_valid, y_valid)],
          verbose = False)


```

```python
import matplotlib.pyplot as plt
print(model.feature_importances_)
# plot
plt.bar(range(len(model.feature_importances_)), model.feature_importances_)
plt.show()
print(X_train.columns)
```

```python
# 預測值
y_pred=model.predict(X_valid)
# 真實值 賦值
y_true= y_valid
print("Accuracy : %.4g" % metrics.accuracy_score(y_true, y_pred))
print(metrics.precision_recall_fscore_support(y_true, y_pred))
print(metrics.f1_score(y_true, y_pred))
print(metrics.classification_report(y_true, y_pred))
```

### 關於上面我們可以有兩種操作
- 根據每個時間單位的詐騙總量做出排名，排名月靠前，該時間單位發生詐騙的次數可能越多
- 計算 每個時間單位發生的詐騙總數/ 完整時間區間內的詐騙總數，數值越大表示該月分發生的次數越多


```python
data_new
```

```python
ind_col = 'purchase_time_month'

def rank_col(data, ind_col):
    df = data.copy()
    df = pd.crosstab(data[ind_col], data['class'])

    lis = list(range(1, len(df)+1))
    df = df.sort_values(by = 1, ascending=False)

    a = list(df.index)
    data['Rank_'+ind_col] = None

    for i in range(len(df)):
        data.loc[data[ind_col] == a[i], 'Rank_'+ind_col] = lis[i]

# dt = data.copy()
# rank_col(dt, 'purchase_time_month')
# dt

time_lis = ['purchase_time_month', 'purchase_time_day', 'purchase_time_hour',
            'signup_time_month', 'signup_time_day', 'signup_time_hour']

for i in time_lis:
    rank_col(data_new, i)
data_new
```

```python
ind_col = 'purchase_time_month'

def rate_col(data, ind_col):

    df = data.copy()
    df = pd.crosstab(data[ind_col], data['class'])

    df['sum'] = df[0]+ df[1]
    df['rate'] = df[1]/ df['sum']
    
    lis = list(df['rate'])

    a = list(df.index)
    data['rate_'+ind_col] = 0

    for i in range(len(df)):
        data.loc[data[ind_col] == a[i], 'rate_'+ind_col] = lis[i]
    
# dt = data.copy()
# rate_col(dt, 'purchase_time_month')
# dt

time_lis = ['purchase_time_month', 'purchase_time_day', 'purchase_time_hour',
            'signup_time_month', 'signup_time_day', 'signup_time_hour']

for i in time_lis:
    rate_col(data_new, i)
data_new.head()
```

```python
def cross_bar_plot_new(data, base_col, lis):
    fig, saxis = plt.subplots(2, 3,figsize=(16,12))
    a = 0
    b = 0

    for i in lis:
        pd.crosstab(data[i], data[base_col]).plot.bar(ax = saxis[a,b])
        if b<2:
            b+=1
        else:
            b = 0
            a = 1

lis_rank = ['Rank_purchase_time_month', 'Rank_purchase_time_day',
            'Rank_purchase_time_hour', 'Rank_signup_time_month', 
            'Rank_signup_time_day', 'Rank_signup_time_hour']

df = data_new[data_new['class'] == 1]
cross_bar_plot_new(df, 'class', lis_rank)
```

```python
from sklearn import metrics
from xgboost.sklearn import XGBClassifier

model = XGBClassifier(max_depth= 8, eta = 0.3, n_estimators = 1000, learning_rate = 0.005,
                          tree_method = 'gpu_hist', predictor = 'gpu_predictor',
                          random_state =12, scale_pos_weight = 9.738255033557047,
                          reg_lambda = 1, reg_alpha = 0
                         )

# 模型 訓練

model.fit(X=X_train,y=y_train, eval_set=[(X_train, y_train), (X_valid, y_valid)],
          verbose = False)


```
